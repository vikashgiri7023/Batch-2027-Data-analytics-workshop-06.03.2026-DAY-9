```python
# Q1
import pandas as pd

# Load dataset
df = pd.read_csv("students_data.csv")

print("First 5 Records:")
print(df.head())

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Q2
from sklearn.preprocessing import MinMaxScaler, StandardScaler

numeric_cols = df.select_dtypes(include=['int64','float64'])

# Min-Max Normalization
minmax = MinMaxScaler()
df_minmax = pd.DataFrame(minmax.fit_transform(numeric_cols), columns=numeric_cols.columns)

# Standardization
standard = StandardScaler()
df_standard = pd.DataFrame(standard.fit_transform(numeric_cols), columns=numeric_cols.columns)

print("Min-Max Normalized Data:")
print(df_minmax.head())

print("\nStandardized Data:")
print(df_standard.head())
```

```python
# Q3
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score

# Load dataset
df = pd.read_csv("students_data.csv")

# Drop non-numeric column
df = df.drop("Name", axis=1)

# Encode categorical columns
le = LabelEncoder()
df["Gender"] = le.fit_transform(df["Gender"])
df["Result"] = le.fit_transform(df["Result"])

# Features and target
X = df.drop("Result", axis=1)
y = df["Result"]

# Train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# LDA model
lda = LinearDiscriminantAnalysis()
lda.fit(X_train, y_train)

# Prediction
y_pred = lda.predict(X_test)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Q4
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO

# Website URL
url = "https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)"

# Browser headers
headers = {"User-Agent": "Mozilla/5.0"}

# Send request
response = requests.get(url, headers=headers)

# Parse HTML
soup = BeautifulSoup(response.text, "html.parser")

# Find all tables with class "wikitable"
tables = soup.find_all("table", {"class": "wikitable"})

# Take the first proper data table
table_html = str(tables[0])

# Convert to DataFrame using StringIO to avoid FutureWarning
df = pd.read_html(StringIO(table_html))[0]

# Display first 5 rows
print(df.head())
```

```python
# Q6
import pandas as pd

# Sample dataframes
sales = pd.DataFrame({
    "OrderID":[1,2,3,4,5],
    "CustomerID":[101,102,103,104,101],
    "Amount":[250,450,300,500,200],
    "City":["Delhi","Mumbai","Delhi","Pune","Delhi"]
})

customers = pd.DataFrame({
    "CustomerID":[101,102,103,104],
    "Name":["Aman","Riya","Rahul","Neha"],
    "Gender":["M","F","M","F"]
})

# Merge datasets on CustomerID
merged = pd.merge(sales, customers, on="CustomerID")

print("Merged Dataset:")
print(merged)

# Filter: Only Amount > 250
filtered = merged[merged["Amount"] > 250]

print("\nFiltered Dataset (Amount > 250):")
print(filtered)

# Group by City and calculate total amount
grouped = filtered.groupby("City")["Amount"].sum().reset_index()

print("\nGrouped and Aggregated (Total Amount per City):")
print(grouped)
```
